# KMeans Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=1000,centers=3,n_features=2,random_state=1)
from sklearn.cluster import KMeans
```

```python
X
```

**In Unsupervised Learning We will not use y**

```python
y
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
```

*We have to actually make a model which can identify these three clusters.*

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
y
```

**How to make clusters?**

```python
from sklearn.cluster import KMeans
kMeans =  KMeans(n_clusters = 3,random_state=1)
```

```python
kMeans.fit(X_train) # Fit-predict can be done both together
```

```python
y_labels =kMeans.predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1], c = y_labels)
```

# Heirarchical Clustering

```python
from sklearn import datasets

data = datasets.load_iris()
```

```python
X= data.data
Y= data.target
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c = Y)
```

```python
# Agglomerative clustering >> Make Dendogram
```

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title('Dendogram')
sc.dendrogram(sc.linkage(pca_scaled,method ="ward"))
plt.xlabel("The Sample Data")
plt.ylabel("Euclidean Distance")
```

**Aglomerative Clustering Building**

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=4,metric = "euclidean", linkage='ward') # Removed affinity='euclidean'
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
silhouette_coeff = []
for k in range(2, 10):
    agglo = AgglomerativeClustering(n_clusters=k, metric = "euclidean", linkage = "ward")
    agglo.fit(pca_scaled)
    score = silhouette_score(pca_scaled, agglo.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_score
```

```python
silhouette_coeff
```
